### 1.1 Interpolation 2D

On commence par travailler dans les coordonnées locales : 

$$ \begin{pmatrix} u \\ v \end{pmatrix} = N q = \begin{pmatrix} N_1 u_1 + N_2 u_2 + N_3 u_3 \\ N_1 v_1 + N_2 v_2 + N_3 v_3 \end{pmatrix}$$

On utilise les valeurs au extremité de l'élément pour reconstituer le déplacement à l'intérieur de l'élément.

$$ q = \begin{pmatrix} u_1\\ v_1 \\ u_2 \\ v_2 \\ u_3 \\ v_3 \end{pmatrix} $$
vecteur constant

$$ N = \begin{pmatrix} N_1 & 0 & N_2 & 0 & N_3 & 0 \\ 
0 & N_1 & 0 & N_2 & 0 & N_3 \end{pmatrix} $$

avec $$ N_1(X_1, X_2) = 1 - X_1 - X_2 $$
$$ N_2(X_1, X_2) = X_1 $$
$$ N_3(X_1, X_2) = X_2 $$

$$ \varepsilon = \begin{pmatrix} \varepsilon_{11} \\
\varepsilon_{22} \\
\varepsilon_{12} \end{pmatrix} = \begin{pmatrix} \frac{\partial u}{\partial X_1} \\
\frac{\partial v}{\partial X_2} \\
\frac{\partial u}{\partial X_2} + \frac{\partial v}{\partial X_1} \end{pmatrix} 
= \begin{pmatrix} \frac{\partial N_1}{\partial X_1} & 0 & \frac{\partial N_2}{\partial X_1} & 0 & \frac{\partial N_3}{\partial X_1} & 0 \\ 
0 & \frac{\partial N_1}{\partial X_2} & 0 & \frac{\partial N_2}{\partial X_2} & 0 & \frac{\partial N_3}{\partial X_2} \\ 
\frac{\partial N_1}{\partial X_1} & \frac{\partial N_1}{\partial X_2} & \frac{\partial N_2}{\partial X_1} & \frac{\partial N_2}{\partial X_2} & \frac{\partial N_3}{\partial X_1} & \frac{\partial N_3}{\partial X_2}
\end{pmatrix}  \begin{pmatrix} u_1\\ v_1 \\ u_2 \\ v_2 \\ u_3 \\ v_3 \end{pmatrix}  = Bq $$

Ensuite, on utilise la loi de Hooke pour les matériaux linéaires élastiques isotropes pour calculer les contraintes à partir des déformations : 

$$ \sigma = \begin{pmatrix} \sigma_{11} \\
\sigma_{22} \\
\sigma_{12} \end{pmatrix} = C \varepsilon = CBq
$$

avec 
$$ C = \frac{E}{1-\nu^2}\begin{pmatrix} 1 & -\nu & 0 \\ 
-\nu & 1 & 0 \\
0 & 0 & \frac{1}{1-2\nu}
\end{pmatrix}  
$$

### 1.2 Visualisation 2D

```python
# Program to plot 2-D Heat map
# using matplotlib.pyplot.imshow() method
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
  
data = np.arange(100 * 100).reshape(100, 100)
plt.imshow( data , cmap = 'RdBu' , interpolation = 'nearest' )
#cmap : autumn
plt.colorbar()
plt.title( "2-D Heat Map" )
plt.show()

```

```python
data = np.random.random(( 100 , 100 ))
ax = sns.heatmap(data, linewidth=0)
plt.show()
```

```python
def heatmap2d(arr: np.ndarray):
    plt.imshow(arr, cmap='viridis')
    plt.colorbar()
    plt.show()

test_array = np.arange(100 * 50).reshape(100, 50)
heatmap2d(test_array)
```
