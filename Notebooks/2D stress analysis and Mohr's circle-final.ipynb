### Lecture 3: Visualising the Transformation Equations
Below we're going to evaluate the transformation equations for a full range of angles from -180 to +180 degrees. We then plot the three transformation equations to visualise the variation of stress with angle.

```python
#Define stresses on a mutually perpendicular pair of faces
sig_x = 1 #(N/mm^2) Let 1 represent the magnitude of normal stress on the x-face
sig_y = 0.25*sig_x #(N/mm^2 )Normal stress on the y-face
tau_xy = 0.75*sig_x #(N/mm^2) Shear stress on the xy face

#Define a full range of angles
theta = np.arange(-180,181) #(degrees) Range of angles
theta_rad = math.pi/180*theta #(radians) Convert angles to radians

#Evaluate the Transformation equations for the full range of angles
sig_x1 = 0.5*(sig_x+sig_y) + 0.5*(sig_x-sig_y)*np.cos(2*theta_rad) + tau_xy*np.sin(2*theta_rad)
sig_y1 = 0.5*(sig_x+sig_y) - 0.5*(sig_x-sig_y)*np.cos(2*theta_rad) - tau_xy*np.sin(2*theta_rad)
tau_x1y1 = -0.5*(sig_x-sig_y)*np.sin(2*theta_rad) + tau_xy*(np.cos(2*theta_rad))

#Set up figure and axes
fig = plt.figure() 
axes = fig.add_axes([0.1,0.1,2,1]) 

#Plot transformation equations
axes.plot(theta,sig_x1,'royalblue', label="Normal Stress, $\sigma_x$")  
axes.plot(theta,sig_y1, 'seagreen', label="Normal Stress, $\sigma_y$")  
axes.plot(theta,tau_x1y1, 'lightcoral', label="Shear Stress, $t\_{x1y1}$")  

#Plot some guide lines
axes.plot([0,0],[-1.5,1.5],'orange', linestyle="--") #Vertical guide line at 0 degrees
axes.plot([90,90],[-1.5,1.5],'orange', linestyle="--") #Vertical guide line at 90 degrees

#Housekeeping
axes.set_xlabel('Angle of face (degrees)')
axes.set_ylabel('Stress $(N/mm^2)$')
axes.set_title('Stress variation with angle')
axes.set_xlim([-180,180])
axes.set_xticks(np.arange(-180,180,20))
axes.legend(loc='lower left')
plt.grid()
plt.show()
```

### Lecture 4: Principal Stresses and Principal Planes
Below we calculate the principal stresses and principal planes before plotting them on top of our plots of the transformation equations.

```python
#Define stresses on a mutually perpendicular pair of faces
sig_x = 1 #(N/mm^2) Let 1 represent the magnitude of normal stress on the x-face
sig_y = 0.25*sig_x #(N/mm^2 )Normal stress on the y-face
tau_xy = 0.75*sig_x #(N/mm^2) Shear stress on the xy face

#Calculate the Principal stresses
sig_avg = 0.5*(sig_x+sig_y)
H = math.sqrt(((sig_x-sig_y)/2)**2 + (tau_xy)**2)
sig_1 = sig_avg + H
sig_2 = sig_avg - H

#Calculate the angles of the Principal planes
theta_1 = 0.5*math.atan((2*tau_xy)/(sig_x-sig_y)) #Principal angle in radians
theta_1d = theta_1*(180/math.pi) #Principal angle in degrees 
theta_2d = theta_1d + 90 #Second principal angle in degrees

#Generate some output statements
print("- Principal angle, theta_pa is {one} degrees".format(one=round(theta_1d,1)))
print("- Principal angle, theta_pb is {one} degrees".format(one=round(theta_2d,1)))
print("- The maximum principal stress is {one} N/mm^2".format(one=round(sig_1,2)))
print("- The minimum principal stress is {one} N/mm^2".format(one=round(sig_2,2)))

#Define a full range of angles
theta = np.arange(-180,181) #(degrees) Range of angles
theta_rad = math.pi/180*theta #(radians) Convert angles to radians

#Evaluate the Transformation equations for the full range of angles
sig_x1 = 0.5*(sig_x+sig_y) + 0.5*(sig_x-sig_y)*np.cos(2*theta_rad) + tau_xy*np.sin(2*theta_rad)
sig_y1 = 0.5*(sig_x+sig_y) - 0.5*(sig_x-sig_y)*np.cos(2*theta_rad) - tau_xy*np.sin(2*theta_rad)
tau_x1y1 = -0.5*(sig_x-sig_y)*np.sin(2*theta_rad) + tau_xy*(np.cos(2*theta_rad))

#Set up figure and axes
fig = plt.figure() 
axes = fig.add_axes([0.1,0.1,2,1]) 

#Plot transformation equations
axes.plot(theta,sig_x1,'royalblue', label="Normal Stress, $\sigma_x$")  
axes.plot(theta,sig_y1, 'seagreen', label="Normal Stress, $\sigma_y$")  
axes.plot(theta,tau_x1y1, 'lightcoral', label="Shear Stress, $t\_{x1y1}$")  

#Guidelines for principal stresses
axes.plot([-180,180],[sig_1,sig_1], 'cyan', linestyle="--") #Horizontal guide line at max principal stress
axes.plot([-180,180],[sig_2,sig_2], 'lime', linestyle="--") #Horizontal guide line at max principal stress
axes.plot([theta_1d,theta_1d],[-1.5,1.5], 'red', linestyle="--") #Vertical guide line at principal plane
axes.plot([theta_2d,theta_2d],[-1.5,1.5], 'red', linestyle="--") #Vertical guide line at principal plane

#Housekeeping
axes.set_xlabel('Angle of face (degrees)')
axes.set_ylabel('Stress $(N/mm^2)$')
axes.set_title('Stress variation with angle')
axes.set_xlim([-180,180])
axes.set_xticks(np.arange(-180,180,20))
axes.legend(loc='lower left')
plt.grid()
plt.show()
```

### Lecture 7: Worked Example - coded solution
In the code below, we develop a coded version of the solution to the example we solved in lecture 6. 

```python
#Define question input parameters
sig_x = -32.5 #(N/mm^2)
sig_y = 8.2 #(N/mm^2)
tau_xy = -10 #(N/mm^2)
theta_d = -35 #(degrees) - Clockwise rotation
theta = theta_d*math.pi/180 #(rads)

# (i) Stresses on element orientated at theta degrees
sig_x1 = 0.5*(sig_x+sig_y) + 0.5*(sig_x-sig_y)*np.cos(2*theta) + tau_xy*np.sin(2*theta)
sig_y1 = 0.5*(sig_x+sig_y) - 0.5*(sig_x-sig_y)*np.cos(2*theta) - tau_xy*np.sin(2*theta)
tau_x1y1 = -0.5*(sig_x-sig_y)*np.sin(2*theta) + tau_xy*(np.cos(2*theta))

print('(i) Stresses on planes orientated at {one} degrees to the xy axes'.format(one=theta_d))
print('- sig_x1 = {one} N/mm^2'.format(one=round(sig_x1,1)))
print('- sig_y1 = {one} N/mm^2'.format(one=round(sig_y1,1)))
print('- tau_x1y1 = {one} N/mm^2'.format(one=round(tau_x1y1,1)))
print('')

#(ii) Principal planes
theta_1 = 0.5*math.atan((2*tau_xy)/(sig_x-sig_y)) #Principal angle in radians
theta_2 = theta_1+(math.pi/2)
theta_1d = theta_1*(180/math.pi) #Principal angle in degrees 
theta_2d = theta_2*(180/math.pi) #Principal angle in degrees 

print('(ii) Principal angles')
print('- The principal angles are {one} and {two} degrees'.format(one=round(theta_1d,1), two=round(theta_2d,1)))
print('')

#(iii) Principal stresses
sig_1 = 0.5*(sig_x+sig_y) + 0.5*(sig_x-sig_y)*np.cos(2*theta_1) + tau_xy*np.sin(2*theta_1) #theta_1d
sig_2 = 0.5*(sig_x+sig_y) + 0.5*(sig_x-sig_y)*np.cos(2*theta_2) + tau_xy*np.sin(2*theta_2) #theta_2d
print('(iii) Principal stresses')
print('- The principal stress {one} N/mm^2 occurs on a plane orientated at {two} degrees'.format(one=round(sig_1,1), two=round(theta_1d,1)))
print('- The principal stress {one} N/mm^2 occurs on a plane orientated at {two} degrees'.format(one=round(sig_2,1), two=round(theta_2d,1)))

if(sig_1>sig_2):
    print('- sig_1 is {one} N/mm^2 and theta_p1 is {two} degrees'.format(one=round(sig_1,1),two=round(theta_1d)))
    print('- sig_2 is {one} N/mm^2 and theta_p2 is {two} degrees'.format(one=round(sig_2,1),two=round(theta_2d)))
else:    
    print('- sig_1 is {one} N/mm^2 and theta_p1 is {two} degrees'.format(one=round(sig_2,1),two=round(theta_2d)))
    print('- sig_2 is {one} N/mm^2 and theta_p2 is {two} degrees'.format(one=round(sig_1,1),two=round(theta_1d)))
print('')


#(iv) Planes of max +ve/-ve shear stress
if(sig_1>sig_2):
    theta_s1 = theta_1d-45 #Plane for maximum positive shear stress    
else:
    theta_s1 = theta_2d-45 #Plane for maximum positive shear stress

theta_s2 = theta_s1+90 #Plane for maximum negative shear stress

print('(iv) Planes of max +ve/-ve shear stress')
print('- Max +ve shear stress occurs on a plane at {one} degrees'.format(one=round(theta_s1,1)))
print('- Max -ve shear stress occurs on a plane at {one} degrees'.format(one=round(theta_s2,1)))
print('')

#(v) Maximum shear stress
tau_max = abs(0.5*(sig_2-sig_1)) #(N/mm^2) Magnitude of maximum shear stress
sig_s = 0.5*(sig_x+sig_y)
print("(v) Max Shear stress")
print('- The maximum shear stress has a magnitude of {one} N/mm^2'.format(one=round(tau_max,1)))
print('- The normal stress on the max shear stress planes is {one} N/mm^2'.format(one=round(sig_s,1)))

#Define a full range of angles
Theta = np.arange(-180,181) #(degrees) Range of angles in degrees
Theta_rad = math.pi/180*Theta #(radians) Convert angles to radians

#Evaluate the Transformation equations for the full range of angles
Sig_x1 = 0.5*(sig_x+sig_y) + 0.5*(sig_x-sig_y)*np.cos(2*Theta_rad) + tau_xy*np.sin(2*Theta_rad)
Sig_y1 = 0.5*(sig_x+sig_y) - 0.5*(sig_x-sig_y)*np.cos(2*Theta_rad) - tau_xy*np.sin(2*Theta_rad)
Tau_x1y1 = -0.5*(sig_x-sig_y)*np.sin(2*Theta_rad) + tau_xy*(np.cos(2*Theta_rad))

#Set up figure and axes
fig = plt.figure() 
axes = fig.add_axes([0.1,0.1,2,1]) 

#Plot transformation equations
axes.plot(Theta,Sig_x1,'royalblue', label="Normal Stress, $\sigma_x$")
axes.plot(Theta,Sig_y1, 'seagreen', label="Normal Stress, $\sigma_y$")  
axes.plot(Theta,Tau_x1y1, 'lightcoral', label="Shear Stress, $t\_{x1y1}$")  

#Plot stresses at 35 degrees
axes.plot([theta_d,theta_d],[sig_2,sig_1], 'r', linestyle="--", lw=2) #Vertical guide line at 35 degrees
axes.plot([theta_d],[sig_x1],'o', color='lime', markersize=5) #Stress sig_x1
axes.plot([theta_d],[sig_y1],'o', color='lime', markersize=5) #Stress sig_y1
axes.plot([theta_d],[tau_x1y1],'o', color='lime', markersize=5) #Stress tau_x1y1

#Plot principal stress and principal planes
axes.plot([theta_1d,theta_1d],[sig_2,sig_1], 'r', linestyle="--", lw=2) #Vertical guide line for principal stress sig_2 (smaller)
axes.plot([theta_2d,theta_2d],[sig_2,sig_1], 'r', linestyle="--", lw=2) #Vertical guide line for principal stress sig_1 (larger)
axes.plot([theta_1d],[sig_1],'o', color='lime', markersize=5) #Principal stress sig_2 (smaller)
axes.plot([theta_2d],[sig_2],'o', color='lime', markersize=5) #Principal stress sig_1 (larger)

#Plot max +ve/-ve shear stress
axes.plot([theta_s1,theta_s1],[-tau_max,tau_max], 'r', linestyle="--", lw=2) #Vertical guide line for max +ve shear stress
axes.plot([theta_s2,theta_s2],[-tau_max,tau_max], 'r', linestyle="--", lw=2) #Vertical guide line for max -ve shear stress
axes.plot([theta_s1],[tau_max],'o', color='lime', markersize=5) #The maximum positive shear stress
axes.plot([theta_s2],[-tau_max],'o', color='lime', markersize=5) #The maximum negative shear stress
axes.plot([theta_s1],[sig_s],'o', color='lime', markersize=5) #The normal stress on the plane of max positive shear stress
axes.plot([theta_s2],[sig_s],'o', color='lime', markersize=5) #The normal stress on the plane of max negative shear stress

#Housekeeping
axes.set_xlabel('Angle of face (degrees)')
axes.set_ylabel('Stress $(N/mm^2)$')
axes.set_title('Stress variation with angle')
axes.set_xlim([-180,180])
axes.set_xticks(np.arange(-180,180,20))
axes.legend(loc='lower left')
plt.grid()
plt.show()
```

### Lecture 8: Mohr's Circle derivation assistance
In the code block below we make use of Python's symbolic math library, SymPy. We use this library to handle the algebra associated with one of the steps of our Mohr's circle derivation. This isn't essential but gives a good demonstration of how to deploy Python to solve a symbolic math problem. 

```python
import sympy as sym #Import SymPy

#Define some symbols 
sX = sym.Symbol('sx') #sigma_x
sY = sym.Symbol('sy') #sigma_y
tXY = sym.Symbol('txy') #tau_xy
T = sym.Symbol('T') #theta

expA = ((sX-sY)/2)*sym.cos(2*T) + tXY*sym.sin(2*T)
expB = -((sX-sY)/2)*sym.sin(2*T) + tXY*sym.cos(2*T)
expC = expA**2 + expB**2 
print(sym.simplify(expC))
```

### Lecture 9: Mohr's Circle example
In the code block below we plot Mohr's circle for the example questions we covered in lectures 6 and 7. Our objective in this code block is simply to plot Mohr's circle and the export it for further discussion in our notepad. 

```python
#Define question input parameters
sig_x = -32.5 #(N/mm^2)
sig_y = 8.2 #(N/mm^2)
tau_xy = -10 #(N/mm^2)

#Construct the key circle parameters
sig_avg = (sig_x+sig_y)/2 #(N/mm^2) Average normal stress 
R = math.sqrt(((sig_x-sig_y)/(2))**2 + (tau_xy)**2) #(N/mm^2) Radius of the circle

#Define a full range of angles
theta = np.arange(0,361) #(degrees) Range of angles in degrees
theta_rad = math.pi/180*theta #(radians) Convert angles to radians

#Generate the cartesian coordinates of the circle for plotting
sig_x1 = sig_avg + R*np.cos(theta_rad) #Cartesian 'x' coordinate
tau_x1y1 = R*np.sin(theta_rad) #Cartesian 'y' coordinate

#Set up figure and axes
fig = plt.figure() 
axes = fig.add_axes([0.1,0.1,2,2]) 
fig.gca().set_aspect('equal', adjustable='box') #Enforce an equal aspect ratio on the plot

#Plot the circle
axes.plot([sig_avg,sig_avg],[-R,R],'--', color='red', linewidth=1) #Plot vertical diameter
axes.plot([sig_avg-R,sig_avg+R],[0,0],'--', color='red', linewidth=1) #Plot horizontal diameter
axes.plot([sig_avg],[0],'o',color='black') #Plot centre of circle
axes.plot(sig_x1,tau_x1y1) #Plot Mohr's Circle

#Plot point 'A' - stress on positive x-face of element
axes.plot([sig_x],[tau_xy],'o',markerfacecolor='g', markeredgecolor='black') #Plot point on the circle
axes.plot([sig_x, sig_avg],[tau_xy,0],'g--' ) #Plot a radial line

#Plot point 'B' - stress on positive y-face of element
axes.plot([sig_y],[-tau_xy],'o',markerfacecolor='orange', markeredgecolor='black') #Plot point on the circle
axes.plot([sig_y, sig_avg],[-tau_xy,0],'--', color='orange' ) #Plot a radial line

#Plot Principal stresses on the circle
axes.plot([sig_avg+R],[0],'X',markerfacecolor='red', markeredgecolor='black', markersize=10) #Plot the major principal stress
axes.plot([sig_avg-R],[0],'X',markerfacecolor='red', markeredgecolor='black', markersize=10) #Plot the minor principal stress

#Plot Max +ve/-ve shear stresses on the circle
axes.plot([sig_avg],[R],'D',markerfacecolor='red', markeredgecolor='black', markersize=10) #Plot the max +ve shear stress
axes.plot([sig_avg],[-R],'D',markerfacecolor='red', markeredgecolor='black', markersize=10) #Plot the max -ve shear stress

#Housekeeping
axes.invert_yaxis()
axes.set_xlabel('Normal stress $\sigma_{x1}$ $(N/mm^2)$')
axes.set_ylabel('Shear stress t$_{x1,y1} (N/mm^2)$')
axes.set_title('Stress variation with angle')
axes.grid()
plt.savefig('Mohr.png',dpi=100, bbox_inches='tight') #Save out the figure to same folder as this Jupyter notebook
plt.show()
```
