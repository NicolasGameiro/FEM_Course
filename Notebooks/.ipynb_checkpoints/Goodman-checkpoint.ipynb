# Fatigue Analysis

### Goodman diagram

On trace le diagramme de goodman pour un matériau donné : 

![modelisation schema](..//Images/Goodman.jpg "title")

```python
Sy = 250 # Max alternative Stress (Endurance Limit) [en MPa]
Sx = 500 # Max mean Stress (Yield Limit) [en MPa]
S = (300,150) # Operating point

import matplotlib.pyplot as plt

plt.plot([0,Sx],[250,0],color='blue',label="Goodman Curve") #Goodman curve
plt.plot(300,150,color='red',marker="*",label='Operating Point')
plt.legend()
```
