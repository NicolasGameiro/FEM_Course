Il faut vérifier que les éléments ne sont pas trop déformer sinon cela peut conduire à des calculs de jacobinne négative ce qui faussera les calculs. On peut définir une paramètre qui s'appelle l'"Aspect Ratio", le "Ratio d'aspect" en français qui permet d'évaluer rapidement la forme des éléments. 

* https://www.youtube.com/watch?v=Vg_iDSXdP_k : mesh quality examples
* https://www.youtube.com/watch?v=v9CFbg87oZA : mesh tool in catia example on hole plate
* https://www.youtube.com/watch?v=kasvCYVAPFE : Adaptative mesh (photo)

```python
import matplotlib.pyplot as plt
import pygmsh

with pygmsh.geo.Geometry() as geom:
    geom.add_polygon(
        [
            [0.0, 0.0],
            [1.0, 0.0],
            [1.0, 1.0],
            [0.0, 1.0],
        ],
        mesh_size=0.2,
    )
    mesh = geom.generate_mesh()

x,y,z = [], [], []
for i in range(len(mesh.points)):
    x.append(mesh.points[i][0])
    y.append(mesh.points[i][1])
    z.append(mesh.points[i][2])

mesh.cells
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x,y,z, c='r', marker='o')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
```
